```python
import pandas as pd
import ast
```

```python
# Read in the raw data
df = pd.read_csv('./data/cleaned2.csv', low_memory=False)
len(df)
```

```python
#Code to convert csv string columns back to lists 

df['categories'] = df['categories']\
    .apply(lambda x: ast.literal_eval(x))
df['mechanics'] = df['mechanics']\
    .apply(lambda x: ast.literal_eval(x))
```

```python
#Code to parse poll column

def parse_poll(poll, key):
    poll_list = ast.literal_eval(poll)
    if isinstance(poll_list, dict):
        if poll_list['@name'] == key:
            return [poll_list['results']["result"]]
        return pd.NA
    if isinstance(poll_list, list):
        try:
            filtered = list(
                filter(lambda x: x['@name'] == key,
                poll_list
            ))
            return list(map(lambda x: x['results']['result'], filtered))
        except:
            print(poll_list) 
            print(filtered)
            return pd.NA
        return pd.NA

df['language_poll'] = df['poll'].apply(lambda x: parse_poll(x, 'language_dependence'))
```

```python
#isolating the id and language column in its own df
language_df = df[["@id","language_poll"]]
language_df
```

```python
#dropping rows that do not contain language poll info (NA's)
lan_df = language_df.dropna()
lan_df
```

```python
#parsing out the language poll column and converting dictionaries to columns
lan_df_exploded = lan_df.explode('language_poll')['language_poll'].apply(pd.Series)

# Concatenate the exploded DataFrame with the original DataFrame
# lan_df_with_dictionaries = pd.concat([lan_df[['@id']], lan_df_exploded], axis=1)
```

```python
lan_df_exploded
```

```python
#renaming the exploded columns to the related @value for that column

lang_df = lan_df_exploded.rename(columns = {0:"Not Dependent", 1:"Somewhat", 2:"Moderately", 3:"Extensively", 4:"Fully"})
display(lang_df.head(2))

lang_df.columns

```

```python
#converting the string values into integers, updating the values to contain only the @numvotes values

def extract_str_conv_int (d, key):
    if key in d:
        return int(d[key])
    else: 
        return None
  
for col in lang_df.columns:
    lang_df[col] = lang_df[col].apply(lambda x: extract_str_conv_int (x, '@numvotes'))

lang_df
```

```python
#added a column where the values contain the exploded column name with the largest num vote count. 
lang_df.info()

lang_df["Language Dependency"] = lang_df.idxmax(axis=1)
lang_df
```

```python
# concatanated the new df with the language df that contained the @id column
languages_votecounts_df = pd.concat([lan_df,lang_df], axis = 1)
languages_votecounts_df
```

```python
#new df created that contains only the id and language dependency
languages_df = languages_exploded_df[["@id", "Language Dependency"]].copy()

languages_df
```
